```python
import pandas as pd
```

```python
kdd_df = pd.read_csv('/Users/sawyer/Downloads/kddcup.csv')
kdd_df

```

```python
key_sets = [set() for key in kdd_df]
def add_to_set(row):
    for i, v in enumerate(row):
        key_sets[i].add(v)
kdd_df.apply(add_to_set, axis=1)
```

```python
key_lists = [list(set) for set in key_sets]
for i, l in enumerate(key_lists):
    l.sort()
    l = [str(x) for x in l]
    key_lists[i] = l
for i, key in enumerate(kdd_df):
    print('Feature: ',key)
    print('Possible Outputs: ',",".join(key_lists[i]))
    
```

```python
import matplotlib.pyplot as plt
keys = [key for key in kdd_df]
columnI = -1
counts = {key: 0 for key in key_sets[columnI]}
for v in kdd_df[keys[columnI]]:
    counts[v] += 1
print(keys[columnI])
plt.barh(list(counts), list(counts.values()), )
key_sets[columnI]
```
